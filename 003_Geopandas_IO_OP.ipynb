### Working with Geopandas

#### 2) Understanding the Supported Vector Input and Output Files
Refer:: https://autogis-site.readthedocs.io/en/2020_/notebooks/L2/00-data-io.html

````python
import geopandas as gpd
gpd.options.io_engine = "fiona"
````

````python
import fiona
fiona.supported_drivers
````

````python
# Read Shapefile
state = gpd.read_file("geodata/State.shp")
````

````python
state.plot()
````

````python
# Write to Shapefile (just make a copy)
state.to_file("geodata/State_1.shp")
````

````python
# Read GeoJSON
read_json_path = "geodata/mysore.geojson"
mysore = gpd.read_file(read_json_path, driver="GeoJSON")
````

````python
mysore.plot();
````

````python
# Write to GeoJSON (just make a copy)
write_json_path = "geodata/mysore_1.geojson"
mysore.to_file(write_json_path, driver="GeoJSON")
````

````python
# Read file from KML/KMZ
fiona.drvsupport.supported_drivers['libkml'] = 'rw'
fiona.drvsupport.supported_drivers['LIBKML'] = 'rw'
fp = "geodata/District.kmz"
mykml = gpd.read_file(fp)
````

````python
mykml.plot();
````

````python
# Write to KML (just make a copy)
outfp = "geodata/District_1.kmz"
mykml.to_file(outfp, driver="LIBKML")
````
