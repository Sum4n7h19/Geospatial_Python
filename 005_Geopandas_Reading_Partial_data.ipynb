# Geopandas - Reading Partial Data

```python
# Importing Geopandas
import geopandas as gpd
%matplotlib inline
```

```python
district = gpd.read_file("https://kgis.ksrsac.in/kgisdocuments/PDF_KML_SHP/District/Shapefiles/District.zip")
```

```python
district.plot()
```

```python
district.head()
```

```python
Belagavi = district[0:1]
```

```python
Belagavi
```

```python
Belagavi.plot()
```

```python
# Saving partial data
Belagavi.to_file("belagavi.shp")
```

```python
mysore = district[district["KGISDist_1"] == "Mysuru"]
```

```python
mysore.plot()
```

```python
mysore.to_file('mysuru.shp')
```

```python
# Seperating Banglore(urban) and Banglore(Rural)
bengaluru = ['Bengaluru (Urban)','Bengaluru (Rural)']
district.loc[district['KGISDist_1'].isin(bengaluru)]
```

```python
selected_Dist = district.loc[district['KGISDist_1'].isin(bengaluru)]
selected_Dist.plot()
```

```python
districtList = district["KGISDist_1"].tolist()
```

```python
districtList
```

```python
districtList = list(map(lambda x: x.replace('Bengaluru (Urban)', 'Bengaluru_urban'), districtList))
```

```python
districtList
```

```python
districtList = list(map(lambda x: x.replace('Bengaluru (Rural)', 'Bengaluru_rural'), districtList))
```

```python
districtList
```

```python
districtList = list(map(lambda x: x.replace(' ', '_'), districtList))
```

```python
districtList
```

```python

mysore = district[district["KGISDist_1"] == "Mysuru"]
```

```python
for dist in districtList:
    temp = district[district["KGISDist_1"] == f"{dist}"]
    temp.to_file(f"{dist}.shp")
    print(f"{dist}.shp saved")
```
