```python
# Import pandas and geopandas
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv("Capital_latlong.csv")
```

```python
df
```

<div style='background:#B1F9E3; color:#000000' class= 'alert alert-box'>
A GeoDataFrame needs a shapely object. We use geopandas points_from_xy() to transform Longitude and Latitude into a list of shapely.
Point objects and set it as a geometry while creating the GeoDataFrame. (note that points_from_xy() is an enhanced wrapper for [Point(x, y) for x, y in zip(df.Longitude, df.Latitude)]).<br><br>
The crs value is also set to explicitly state the geometry data defines latitude/ longitude world geodetic degree values. 
This is important for the correct interpretation of the data, such as when plotting with data in other formats.
</div>

```python
capital = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude), crs="EPSG:4326")
```

```python
capital
```

```python
capital.plot()
```

```python
capital.to_file('geodata/capitalhq.shp')
```
