```python
# working with map classification(classifying quantative values)
# Refer: https://geopandas.org/en/stable/gallery/choropleths.html
```

Map Classification by Quantile Method:

```python
layer1 = gdf.boundary.plot(edgecolor='black', linewidth = 0.2, figsize=(8, 8))
customized={"label": "Population in 2011", "orientation": "horizontal"}

# Add Scheme and K value and remove legend_kwds
ax1 = gdf.plot(ax = layer1, column="POP_F",scheme="quantiles", k=3, legend=True, figsize=(8, 8), cmap ='OrRd');

ax1.set_title('Total Female Population of Bengaluru City, 2011')

# Parameters = longitude,latitude,orientation,size
ax1.arrow(77.75, 13.15, 0, 0.01, length_includes_head=True,head_width=0.03, head_length=0.05, overhang=.1, facecolor='k')
```

Map Classification by equal intervals Method:

```python
layer1 = gdf.boundary.plot(edgecolor='black', linewidth = 0.2, figsize=(8, 8))
customized={"label": "Population in 2011", "orientation": "horizontal"}

# Add Scheme and K value and remove legend_kwds
ax1 = gdf.plot(ax = layer1, column="POP_F",scheme="equal_interval", k=3, legend=True, figsize=(8, 8), cmap ='OrRd');

ax1.set_title('Total Female Population of Bengaluru City, 2011')

# Parameters = longitude,latitude,orientation,size
ax1.arrow(77.75, 13.15, 0, 0.01, length_includes_head=True,head_width=0.03, head_length=0.05, overhang=.1, facecolor='k')
```

Map Classificaton by natural breaks Method:

```python
layer1 = gdf.boundary.plot(edgecolor='black', linewidth = 0.2, figsize=(8, 8))
customized={"label": "Population in 2011", "orientation": "horizontal"}

# Add Scheme and K value and remove legend_kwds
ax1 = gdf.plot(ax = layer1, column="POP_F",scheme="natural_breaks", k=3, legend=True, figsize=(8, 8), cmap ='OrRd');

ax1.set_title('Total Female Population of Bengaluru City, 2011')

# Parameters = longitude,latitude,orientation,size
ax1.arrow(77.75, 13.15, 0, 0.01, length_includes_head=True,head_width=0.03, head_length=0.05, overhang=.1, facecolor='k')
```

```python
# For Generating Random Color to Categorical Values
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
vals = np.linspace(0,1,256)
np.random.shuffle(vals)
mycmap = matplotlib.colors.ListedColormap(np.random.rand(256,3))

#-----------------------------------------------------------------------------------------

layer1 = gdf.boundary.plot(edgecolor='black', linewidth = 0.2, figsize=(8, 8))
customized={"label": "Population in 2011", "orientation": "horizontal"}

# Add Catergorical is True and Point CMAP to Custom mycmap
# Move the legend position to required position
ax1 = gdf.plot(ax = layer1,
               column="ASS_CONST1",
               categorical = True,
               legend=True,
               legend_kwds = {'bbox_to_anchor':(1.4,1)},
               figsize=(8, 8),
               cmap = mycmap);

ax1.set_title('Assembly Constitutions of Bengaluru City, 2011')

# Parameters = longitude,latitude,orientation,size
ax1.arrow(77.75,
          13.15,
          0,
          0.01,
          length_includes_head=True,
          head_width=0.03,
          head_length=0.05,
          overhang=.1,
          facecolor='k')
```

```python
ax1.figure.savefig('Assembly_Constitution.jpg', dpi=150, bbox_inches='tight')
```

```python
# Automating Maps

# List of field to Map
maplist  = ['POP_M', 'POP_F','POP_SC','POP_ST','POP_TOTAL']
for field in maplist:
    layer1 = gdf.boundary.plot(edgecolor='black', linewidth = 0.2, figsize=(8, 8))
    customized={"label": "Population in 2011", "orientation": "horizontal"}

    # Add Scheme and K value and remove legend_kwds
    ax1 = gdf.plot(ax = layer1, column=f'{field}',scheme="equal_interval", k=3, legend=True, figsize=(8, 8), cmap ='OrRd');
    
    if f'{field}' == 'POP_M':
        ax1.set_title('Total Male Population of Bengaluru City, 2011')
    elif f'{field}' == 'POP_F':
        ax1.set_title('Total Female Population of Bengaluru City, 2011')
    elif f'{field}' == 'POP_SC':
        ax1.set_title('Total SC Population of Bengaluru City, 2011')
    elif f'{field}' == 'POP_ST':
        ax1.set_title('Total ST Population of Bengaluru City, 2011')
    else:
        ax1.set_title('Total Population of Bengaluru City, 2011')
        
    # Parameters = longitude,latitude,orientation,size
    ax1.arrow(77.75, 13.15, 0, 0.01, length_includes_head=True,head_width=0.03, head_length=0.05, overhang=.1, facecolor='k')
    filename = f'{field}.jpg'
    ax1.figure.savefig(filename, dpi=150, bbox_inches='tight')
    print(f'{filename}, saved')
```
