```python
# Applying Mathematics on Grid Values (NetCDF)
# Currently Grid Contains Temperature in Degree Kelvin
# Question is How can we convert Kelvin to Celcius and Store back to dataset
```

```python
kelvin = ds.air.mean(dim='time')
kelvin.plot();
```

```python
celsius = kelvin - 273.16
celsius.plot();
```

```python
# Adding Data to DataSets
ds['celsius'] = celsius
ds['kelvin'] = kelvin
ds
```

```python
ds.air.attrs
```

```python
ds.celsius.attrs
```

```python
ds.kelvin.attrs
```

```python
ds.kelvin.attrs['Description'] = 'Mean air tempterature (through time) in kelvin.'
ds.celsius.attrs['Description'] = 'Mean air tempterature (through time) in celsius.'
```

```python
ds.kelvin.attrs
```

```python
ds.to_netcdf('multi-dim-data/air_temperature_modified.nc')
```

```python
try:
    ds.air.sel(lat=70.5, lon=260.5).plot();
except:
    print("No Value at this particular point")
```

```python
# Lets solve the above problem to passed nearest neighbour method
ds.air.sel(lat=70.5, lon=260.5, method='nearest', tolerance=5).plot();
```

```python
requiredLocation = ds.air.sel(lat=70.5, lon=260.5, method='nearest', tolerance=5)
```

```python
requiredLocation
```

```python
df = requiredLocation.to_dataframe()
```

```python
import cartopy.crs as ccrs
```

```python
t_average = ds.air.mean(dim='time')

f = plt.figure(figsize=(8, 4))
# Define the map projection *on which* you want to plot
ax = plt.axes(projection=ccrs.Orthographic(-80, 35))
# ax is an empty plot. We now plot the variable t_average onto ax
# the keyword "transform" tells the function in which projection the air temp data is stored 
t_average.plot(ax=ax, transform=ccrs.PlateCarree())
# Add gridlines and coastlines to the plot
ax.coastlines(); ax.gridlines(); 
```
